## DATA MANIPULATION



Calculating the status of the order:

```python
# Load the CSV file into a DataFrame
orders_new = pd.read_csv('orders_new.csv')

# Convert date columns to datetime format
orders_new['ScheduledDeliveryDate'] = pd.to_datetime(orders_new['ScheduledDeliveryDate'])
orders_new['ActualDeliveryDate'] = pd.to_datetime(orders_new['ActualDeliveryDate'])
orders_new['OrderDate'] = pd.to_datetime(orders_new['OrderDate'])

# Calculate the difference between ActualDeliveryDate and ScheduledDeliveryDate
orders_new['DeliveryDelay'] = orders_new['ActualDeliveryDate'] - orders_new['ScheduledDeliveryDate']

# Define a function to categorize the status of the order
def order_status(row):
    if row['DeliveryDelay'] < pd.Timedelta(0):  # Early delivery
        return 'Early Delivery'
    elif row['DeliveryDelay'] > pd.Timedelta(0):  # Late delivery
        return 'Late Delivery'
    else:  # On-time delivery
        return 'On-Time'

# Apply the function to each row to create a new column 'OrderStatus'
orders_new['Status'] = orders_new.apply(order_status, axis=1)
orders_new.to_csv('orders_new.csv', index=False)
# Print the DataFrame with the new column
orders_new.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load customers and city data
customers = pd.read_csv('customers.csv')
city = pd.read_csv('city.csv')

# Merge customers with city to get city names
merged_data = pd.merge(customers, city, on='CityID', how='left')

# Group data by service channel and calculate the number of unique cities
customers_by_channel = merged_data.groupby(['ServiceChannel'])['City'].count().reset_index()
customers_by_channel.columns = ['ServiceChannel', 'Number of Customers']

# Sort service channels by the number of customers and select top 10
top_10_ServiceChannel = customers_by_channel.groupby('ServiceChannel').size().nlargest(10).index
top_10_Customers_by_Channel = customers_by_channel[customers_by_channel['ServiceChannel'].isin(top_10_ServiceChannel)]

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(data=top_10_Customers_by_Channel, x='ServiceChannel', y='Number of Customers', hue='ServiceChannel', palette="magma")
plt.title('Top 10 Service Channels by Number of Customers in Each City')
plt.xlabel('Service Channel')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Load orders_new1.csv
orders_new = pd.read_csv('orders_new.csv')

# Convert 'OrderDate' to datetime format to extract month
orders_new['OrderDate'] = pd.to_datetime(orders_new['OrderDate'])

# Extract month from 'OrderDate'
orders_new['OrderMonth'] = orders_new['OrderDate'].dt.month

# Group data by 'OrderMonth' and count occurrences of each 'Status'
status_count_by_month = orders_new.groupby(['OrderMonth', 'Status']).size().unstack(fill_value=0)

# Set the color palette to magma
sns.set_palette("magma")

# Plotting
status_count_by_month.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Count of Status by Order Month')
plt.xlabel('Order Month')
plt.ylabel('Count of Status')
plt.xticks(rotation=0)
plt.legend(title='Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Load orders_new1.csv
orders_new = pd.read_csv('orders_new.csv')

# Convert 'OrderDate' to datetime format to extract month
orders_new['OrderDate'] = pd.to_datetime(orders_new['OrderDate'])

# Extract month from 'OrderDate'
orders_new['OrderMonth'] = orders_new['OrderDate'].dt.month

# Group data by 'OrderMonth' and count occurrences of each 'Status'
status_count_by_month = orders_new.groupby(['OrderMonth', 'Returns']).size().unstack(fill_value=0)

# Set the color palette to icefire
sns.set_palette("viridis")

# Plotting
status_count_by_month.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Count of Returns by Order Month')
plt.xlabel('Order Month')
plt.ylabel('Count of Returns')
plt.xticks(rotation=0)
plt.legend(title='Returns', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Load data from CSV files
orders_new = pd.read_csv('orders_new.csv')
sales = pd.read_csv('sales.csv')

# Merge orders with sales data to get salesperson details
merged_data = pd.merge(orders_new, sales, on='SalespersonID', how='left')

# Group by Salesperson and Status to count occurrences
status_count = merged_data.groupby(['Salesperson', 'Status']).size().reset_index(name='Count')

# Calculate total orders per salesperson
total_orders = status_count.groupby('Salesperson')['Count'].sum().reset_index()
top_10_salespersons = total_orders.nlargest(10, 'Count')['Salesperson']

# Filter the status_count for only the top 10 salespersons
top_10_status_count = status_count[status_count['Salesperson'].isin(top_10_salespersons)]

# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(data=top_10_status_count, x='Salesperson', y='Count', hue='Status', palette="magma")
plt.title('Count of Order Statuses by Top 10 Salespersons')
plt.xlabel('Salesperson')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.legend(title='Status')
plt.tight_layout()
plt.show()
```

```python
# Load data
orders_new = pd.read_csv('orders_new.csv')
customers_new = pd.read_csv('customers_new.csv')
city = pd.read_csv('city.csv')  # Assuming City information includes a key to join with sales or orders_new

# Merge orders with customer data to get customer details
orders_customer_new = pd.merge(orders_new, customers_new, on='CustomerID', how='left')

# Merge with city data
merged_data = pd.merge(orders_customer_new, city, on='CityID', how='left')

# Group by City and Status to count occurrences
city_status_count = merged_data.groupby(['City', 'Status']).size().reset_index(name='Count')

# Calculate total orders per city
total_orders_per_city = city_status_count.groupby('City')['Count'].sum().reset_index()
top_ten_cities = total_orders_per_city.nlargest(10, 'Count')['City']

# Filter the original grouped data to include only top ten cities
top_cities_status_count = city_status_count[city_status_count['City'].isin(top_ten_cities)]

# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(data=top_cities_status_count, x='City', y='Count', hue='Status', palette="magma")
plt.title('Count of Order Statuses by Top 10 Cities')
plt.xlabel('City')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.legend(title='Status')
plt.tight_layout()
plt.show()
```

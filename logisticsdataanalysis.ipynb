## DATA MANIPULATION



Calculating the status of the order:

```python
# Load the CSV file into a DataFrame
orders_new = pd.read_csv('orders_new.csv')

# Convert date columns to datetime format
orders_new['ScheduledDeliveryDate'] = pd.to_datetime(orders_new['ScheduledDeliveryDate'])
orders_new['ActualDeliveryDate'] = pd.to_datetime(orders_new['ActualDeliveryDate'])
orders_new['OrderDate'] = pd.to_datetime(orders_new['OrderDate'])

# Calculate the difference between ActualDeliveryDate and ScheduledDeliveryDate
orders_new['DeliveryDelay'] = orders_new['ActualDeliveryDate'] - orders_new['ScheduledDeliveryDate']

# Define a function to categorize the status of the order
def order_status(row):
    if row['DeliveryDelay'] < pd.Timedelta(0):  # Early delivery
        return 'Early Delivery'
    elif row['DeliveryDelay'] > pd.Timedelta(0):  # Late delivery
        return 'Late Delivery'
    else:  # On-time delivery
        return 'On-Time'

# Apply the function to each row to create a new column 'OrderStatus'
orders_new['Status'] = orders_new.apply(order_status, axis=1)
orders_new.to_csv('orders_new.csv', index=False)
# Print the DataFrame with the new column
orders_new.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
orders_new = pd.read_csv('orders_new.csv')
customers_new = pd.read_csv('customers_new.csv')
city = pd.read_csv('city.csv')  # Assuming City information includes a key to join with orders_new

# Merge orders with customer data to get customer details
orders_customer_new = pd.merge(orders_new, customers_new, on='CustomerID', how='left')

# Merge with city data
merged_data = pd.merge(orders_customer_new, city, on='CityID', how='left')

# Filter data for on-time deliveries
on_time_deliveries = merged_data[merged_data['Status'] == 'On-Time']  # Updated to 'On-Time'

# Group by City and count unique OrderIDs for on-time deliveries
city_on_time_delivery_count = on_time_deliveries.groupby('City')['OrderID'].nunique().reset_index(name='On-Time Delivery Count')

# Sort cities by the count of on-time deliveries in ascending order
city_on_time_delivery_count_sorted = city_on_time_delivery_count.sort_values(by='On-Time Delivery Count', ascending=False)

# Select only the top five cities
top_five_cities = city_on_time_delivery_count_sorted.head(5)

# Plotting
plt.figure(figsize=(12, 8))
sns.barplot(data=top_five_cities, x='City', y='On-Time Delivery Count', palette="crest")
plt.title('Top 5 Cities by Number of On-Time Deliveries')
plt.xlabel('City')
plt.ylabel('Number of On-Time Deliveries')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import calendar

# Load the orders data
orders_new = pd.read_csv('orders_new.csv')

# Ensure the 'OrderMonth' is extracted properly if not already in the data
# This step assumes your dates are in a column named 'OrderDate' and are in standard date format
orders_new['OrderDate'] = pd.to_datetime(orders_new['OrderDate'])
orders_new['OrderMonth'] = orders_new['OrderDate'].dt.month  # Extract month if not already present

# Convert month numbers to month names
orders_new['MonthName'] = orders_new['OrderMonth'].apply(lambda x: calendar.month_name[x])

# Filter data for early deliveries
on_time_deliveries = orders_new[orders_new['Status'] == 'On-Time']

# Group by MonthName and count unique OrderIDs for early deliveries
monthly_on_time_deliveries = on_time_deliveries.groupby('MonthName')['OrderID'].nunique().reset_index(name='On-Time Count')

# Sort the count of early deliveries and select the top 5 months by their counts
# This requires a custom sort order based on month name position in a year, which we handle separately
month_order = {month: index for index, month in enumerate(calendar.month_name) if month}
monthly_on_time_deliveries['MonthSort'] = monthly_on_time_deliveries['MonthName'].map(month_order)
top_five_months = monthly_on_time_deliveries.sort_values(by=['On-Time Count', 'MonthSort'], ascending=[False, True]).head(5)

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(data=top_five_months, x='MonthName', y='On-Time Count', palette="crest")
plt.title('Top 5 Months by Number of On-Time Deliveries')
plt.xlabel('Month')
plt.ylabel('Number of On-Time Deliveries')
plt.xticks(rotation=45)  # Adjust rotation if needed
plt.tight_layout()
plt.show()
```

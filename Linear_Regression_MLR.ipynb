## Step 1: Data Preprocessing

```python
dataset = pd.read_csv('/content/50_Startups.csv')
X_mlr = df_dummies.drop(['Profit'],axis = 1)
Y_mlr = df_dummies.Profit
```

```python
from sklearn.model_selection import train_test_split

X_train_mlr, X_test_mlr, Y_train_mlr, Y_test_mlr = train_test_split( X_mlr, Y_mlr, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_mlr= sc.fit_transform(X_train_mlr)
X_test_mlr  = sc.transform(X_test_mlr)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train_mlr, Y_train_mlr)
```

```python
Y_pred_mlr = regressor.predict(X_test_mlr)
```

```python
plt.plot(X_train_mlr , Y_train_mlr, color = 'red')
plt.plot(X_train_mlr, regressor.predict(X_train_mlr), color ='blue')
```

```python
plt.plot(X_test_mlr , Y_test_mlr, color = 'red')
plt.plot(X_test_mlr , regressor.predict(X_test_mlr), color ='blue')
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(Y_test_mlr,Y_pred_mlr)
```

```python
print("Train Score", regressor.score(X_train_mlr , Y_train_mlr))
```

```python
print("Test Score", regressor.score(X_test_mlr , Y_test_mlr))
```

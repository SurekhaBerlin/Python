## DATA MANIPULATION



Calculating the status of the order:

```python
# Load the CSV file into a DataFrame
orders = pd.read_csv('orders.csv')

# Convert date columns to datetime format
orders['ScheduledDeliveryDate'] = pd.to_datetime(orders['ScheduledDeliveryDate'])
orders['ActualDeliveryDate'] = pd.to_datetime(orders['ActualDeliveryDate'])
orders['OrderDate'] = pd.to_datetime(orders['OrderDate'])

# Calculate the difference between ActualDeliveryDate and ScheduledDeliveryDate
orders['DeliveryDelay'] = orders['ActualDeliveryDate'] - orders['ScheduledDeliveryDate']

# Define a function to categorize the status of the order
def order_status(row):
    if row['DeliveryDelay'] < pd.Timedelta(0):  # Early delivery
        return 'Early Delivery'
    elif row['DeliveryDelay'] > pd.Timedelta(0):  # Late delivery
        return 'Late Delivery'
    else:  # On-time delivery
        return 'On-Time'

# Apply the function to each row to create a new column 'OrderStatus'
orders['Status'] = orders.apply(order_status, axis=1)

# Print the DataFrame with the new column
orders.head()
```

```python
# Load the CSV file into a DataFrame
orders = pd.read_csv('orders.csv')

# Convert date columns to datetime format
orders['ScheduledDeliveryDate'] = pd.to_datetime(orders['ScheduledDeliveryDate'])
orders['ActualDeliveryDate'] = pd.to_datetime(orders['ActualDeliveryDate'])
orders['OrderDate'] = pd.to_datetime(orders['OrderDate'])

# Calculate the difference between ActualDeliveryDate and ScheduledDeliveryDate
orders['DeliveryDelay'] = orders['ActualDeliveryDate'] - orders['ScheduledDeliveryDate']

# Define a function to categorize the status of the order
def order_status(row):
    if row['DeliveryDelay'] < pd.Timedelta(0):  # Early delivery
        return 'Early Delivery'
    elif row['DeliveryDelay'] > pd.Timedelta(0):  # Late delivery
        return 'Late Delivery'
    else:  # On-time delivery
        return 'On-Time'

# Apply the function to each row to create a new column 'OrderStatus'
orders['Status'] = orders.apply(order_status, axis=1)

# Print the DataFrame with the new column
orders.head()
```
